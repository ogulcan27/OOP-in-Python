# What is OOP ?

Class (sınıf) ve object (nesne) kavramlarına dayanan bir programlama paradigmasıdır (görüş, bakış açısı). Problemlere üretilen çözümlerin nasıl formül haline getirileceğiyle alakalı olan temel bir programlama stilidir.

### < Object (Nesne) >

Durum (state) ve davranış (behavior) hakkında bilgi içeren veri yapısıdır.



Örneğin yukarıdaki görselde olduğu gibi; bir müşteriyi temsil eden bir nesne, belirli bir telefona sahip olabilir. Bunlara ilişkili olarak placeOrder ve cancelOrder gibi davranışlara sahip olabilir.

Örneğin; bir web sitesinde, bir düğmeyi temsil eden bir nesnenin bir etiketi olabilir ve o düğmeye basıldığında bir davranış sergileyebilir.

OOP'nin ayırt edici özelliği, durum ve davranışları birlikte paketlemesidir. İlk örnekteki müşteri verilerini, müşteri eylemlerinden ayrı olarak düşünmek yerine, bir müşteriyi temsil eden tek birim olarak düşünüyoruz. Buna **Encapsulation (Kapsülleme)** denir ve nesne yönelimli programlamanın temel ilkelerinden biridir.

### < Class (Sınıf) >

Olası durumları ve davranışları özetleyen nesneler için bir plan gibidir. Belirli bir türdeki her nesnenin sahip olabileceği olası durumları ve davranışları tanımlar.

### < Class Anatomy : The __init__ Constructor (__init__ Yapıcısı) > 

Class'lar çok fazla veri içeriyorsa, bu sürdürülemez hale gelebilir. **Nesneye veri eklemek** daha iyi bir strateji olacaktır. Python; **initializing (__ init __)** adı verilen, her nesne oluşturulduğunda otomatik olarak çağırılan bir yapıcı kullanmamıza izin verir. 

Aşağıdaki kod bloğunda Customer sınıfı için __ init __ metodunu kullanalım.

```python
class Customer: 
    
    def __init__(self, name):
        self.name = name #1
        print("The __init__ method was called")
        
cust = Customer("Çağlar") # <--  __init__ dolaylı olarak çağırıldı
print(cust.name)
```

+ **1-)** Yöntemin gövdesinde name niteliğini oluşturuyoruz. name niteliğini, name parametresine girilecek argümanlar oluşturacak 
+ Artık Customer sınıfına ait bir nesne oluştururken parantez içinde müşteri adını girebiliriz. Nesne ve  __ init __ yöntemi otomatik olarak çağrılır ve name özniteliği oluşturulur. 

__ init __ metoduna başka bir parametre daha ekleyelim. 

```python
class Customer:
    
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        
cust = Customer(name = "Lara de Silva", balance = 1000)
print(cust.name)
print(cust.balance)
```

+ Artık parantez içinde 2 parametre ile Customer nesnesi oluşturabileceğiz.

__ init __ yapıcısı, öznitelikler için varsayılan değerleri ayarlamak için de işe yarar. Yukarıdaki **balance** argümanını 0 olacak şekilde varsayılan olarak ayarlayalım. 

```python
class Customer:
    
    def __init__(self, name, balance = 0):
        self.name = name
        self.balance = balance
        
cust = Customer("Lara de Silva")
print(cust.name)
print(cust.balance)
```

**balance** değerini belirtmeden bir Customer nesnesi oluşturduk.

**NOT :** Mümkün olduğunca, __ init __ dışında öznitelik tanımlamaktan kaçınmamız lazım. Class'ımız yüzlerce kod satırı uzunluğunda olabilir ve kodumuzu okuyan kişi tüm nitelikleri bulmak için hepsini taramak zorunda kalır. Ayrıca, __ init __ 'teki tüm niteliklerin tanımlanması, nesne oluşturulduğunda, bütün özniteliklerin oluşturulmasını sağlar. Bu nedenle henüz var olmayan bir özniteliğe erişmeye çalışmak konusunda endişelenmemize gerek kalmaz. Bütün bunlar daha okunaklı bir kod yazmış olmamızı sağlar.  

Yukarıda oluşturduğumuz Employee class'ını __ init __ kullanarak oluşturalım.

```python
class Employee: 
    
    def __init__(self, name, salary = 0):
        self.name = name
        if salary > 0:
            self.salary = salary
        elif salary < 0:
            self.salary = 0
            print("Invalid Salary!")

emp = Employee("Korel Rossi", -1000)
print(emp.name)
print(emp.salary)
```

+ Yukarıdaki Employee class'ı ile bunu karşılaştıracak olursak; her bir öznitelik için ayrı ayrı metot oluşturmaya gerek kalmadan __ init __ ile nesne oluşturulduğu anda özniteliklerin belirlendiği bir Employee class'ı oldu. 

Aşağıdaki kod bloğunda, noktanın orijine olan uzaklığını bulan bir class yazalım. 

```python
import numpy as np

class Point:
    """ A point on a 2D plane
    
   Attributes
    ----------
    x : float, default 0.0. The x coordinate of the point        
    y : float, default 0.0. The y coordinate of the point
    """
    def __init__(self, x=0.0, y=0.0):
      self.x = x
      self.y = y
      
    def distance_to_origin(self):
      """Calculate distance from the point to the origin (0,0)"""
      return np.sqrt(self.x ** 2 + self.y ** 2)
    
    def reflect(self, axis):
      """Reflect the point with respect to x or y axis."""
      if axis == "x":
        self.y = - self.y
      elif axis == "y":
        self.x = - self.x
      else:
        print("The argument axis only accepts values 'x' and 'y'!")
        
pt = Point(x=3.0)
pt.reflect("y")
print((pt.x, pt.y))
pt.y = 4.0
print(pt.distance_to_origin())
```

# Inheritance and Polymorphism (Kalıtım ve Polimorfizm)

Kalıtım ve polimorfizm, kodun verimli ve tutarlı bir şekilde yeniden kullanılmasını sağlayan OOP'nin temel kavramlarıdır. Bir sınıftan devralmayı, yöntemleri özelleştirmeyi ve yeniden tanımlamayı öğreneceğiz ve sınıf düzeyindeki veriler ile örnek düzeyindeki veriler arasındaki farkları inceleyeceğiz.

### < Instance and Class Data (Örnek ve Sınıf Verileri) > 

**Bir class'taki tüm nesnelerin ortak özniteliği**

Bir sınıfa ait bütün nesnelerde ortak olan öznitelik, **sınıf düzeyinde global** bir değişkendir. Sınıfa ait olan bütün nesnelerde aynı değere sahiptir. Aşağıdaki kod bloğunda buna bir örnek verelim.  

```python
class Employee:
    
    MIN_SALARY = 30000
    def __init__(self, name, salary):
        self.name = name
        if salary >= Employee.MIN_SALARY:
            self.salary = salary
        else:
            self.salary = Employee.MIN_SALARY
            
emp1 = Employee("TBD", 400000)
print(emp1.MIN_SALARY)
emp2 = Employee("TBD", 600000)
print(emp2.MIN_SALARY)
```

+ Görüleceği üzere Employee sınfına ait emp1 ve emp2 nesnelerinin **salary** öznitelikleri ne kadar farklı olsa da MIN_SALARY öznitelikleri, **class içerisinde global bir değişken** olduğu için her iki nesnede de aynıdır. 

**@classmethod**

Python, **@classmethod** dekoratörünü ve özel bir ilk bağımsız değişken **cls**'yi kullanarak sınıf yöntemlerini de tanımlamanıza izin verir. Sınıf yöntemlerinin ana kullanımı, sınıfın bir örneğini döndüren, ancak **__ init __()** ile aynı kodu kullanmayan yöntemleri tanımlamaktır. @classmethod, sınıfın bir örneği yerine sınıfın kendisi üzerinde çalışan bir sınıfta bir yöntemi tanımlamak için kullanılan bir dekoratördür. Bu yönteme "sınıf yöntemi" denir. Sınıf yöntemleri genellikle bir sınıf için alternatif oluşturucular olarak kullanılır. Normal **__ init __()** yönteminden farklı giriş parametreleri kullanarak sınıfın yeni örneklerini oluşturmak için kullanılabilirler. Sınıf yöntemleri, sınıfın bir örneğinden ziyade sınıfın kendisinde çağrılır.

Örneğin, bir zaman serisi paketi geliştiriyorsunuz ve tarihlerle çalışmak için kendi sınıfınızı, BetterDate'i tanımlamak istiyorsunuz. Sınıfın öznitelikleri yıl, ay ve gün olacaktır. Yıl, ay ve gün değerleri verilen BetterDate nesneleri oluşturan bir yapıcıya sahip olmak istiyorsunuz, ancak aynı zamanda 2020-04-30 gibi dizelerden BetterDate nesneleri de oluşturabilmek istiyorsunuz. Bu örneği aşağıdaki kod bloğunda görelim. 

```python
class BetterDate:    
    # Constructor
    def __init__(self, year, month, day):
      self.year, self.month, self.day = year, month, day
    
    # from_str adlı @classmethod'u tanımlayalım
    @classmethod 
    def from_str(cls, datestr):
        # Dizeyi "-" ile ayıralım ve her parçayı tamsayıya dönüştürelim
        parts = datestr.split("-")
        year, month, day = int(parts[0]), int(parts[1]), int(parts[2])
        # year, month, day = map(int, datestr.split("-")) yazılırsa daha temiz bir kod olur.
        return cls(year, month, day)
        
bd = BetterDate.from_str('2020-04-30')   
print(bd.year)
print(bd.month)
print(bd.day)
```

+ **@classmethod** ile **from_str** metodu, BetterDate sınıfının her nesnesinin ortak bir metodu olmuş oldu. Ortak bir davranış. **@classmethod**'un global olmayan metotlardan farkı; **self** yerine **cls** kullanılmasıdır. **cls**'nin işlevi **self** ile aynıdır. 

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height
        
    def area(self):
        return self.width * self.height
        
    @classmethod
    def from_square(cls, size):
        return cls(size, size)

# create a rectangle using the constructor
rect = Rectangle(4, 5)
print(rect.area()) # prints 20

# create a square using the class method
square = Rectangle.from_square(3)
print(square.area()) # prints 9
```

+ Bu örnekte, bir __init__() yöntemi ve bir field() yöntemi ile bir Rectangle sınıfı tanımladık. Ayrıca eşit genişlik ve yüksekliğe sahip yeni bir Rectangle sınıfı örneği oluşturan from_square() adında bir @classmethod tanımladık.

  Normal yapıcıyı kullanarak bir dikdörtgen oluşturabiliriz ve sınıf yöntemini kullanarak bir kare oluşturabiliriz. Sınıf yönteminin, sınıfın bir örneği yerine Rectangle sınıfının kendisinde çağrıldığına dikkat edin. Sınıf yöntemi çağrıldığında, Rectangle sınıfının eşit genişlik ve yüksekliğe sahip yeni bir örneğini döndürür ve etkili bir şekilde bir kare oluşturur.
  
  Sınıf yöntemleri, bir sınıfın örneklerini oluşturmanın farklı yollarına izin vererek, daha esnek ve modüler kod oluşturmak için kullanışlıdır. Farklı giriş parametreleri veya yapılandırmaları kullanarak bir sınıfın örneklerini oluşturmanız gereken durumlarda özellikle yararlı olabilirler.

### < Class Inheritance (Sınıf Mirası, Kalıtımı) > 

**Inheritance :** Bir sınıfın, başka bir sınıftan türetilmesidir. Class içinde class olarak düşünülebilir. İlk class; parent (üst), ilk class'ın içindeki class; child (alt) class olarak geçer. Alt sınıf, üst sınıftan tüm genel ve korumalı özellikleri ve yöntemleri devralır. Ek olarak, kendi özelliklerine ve yöntemlerine sahip olabilir.



Örneğin, bir bakiye özniteliği ve para çekme yöntemi olan temel bir banka hesabı (Bank Account) sınfımız var. Şirketimizde çeşitli hesap türleri ile çalıştığımızı varsayalım. Örneğin, bir Tasarruf Hesabı (Saving Accounts) ayrıca bir faiz oranına ve faizi hesaplama yöntemine sahiptir. Ama yine de bir dengesi olacak şekilde ondan para çekebilmeliyiz. BankAccount'tan, SavingAccount yöntemlerini ve özelliklerini devralarak BankAccount sınıfı için önceden yazdığımız kodu yeniden kullanabileceğiz. Bakiyesi ve para çekme yöntemi de olan bir CheckingAccount sınıfımız olabilir.

Aşağıdaki kod bloğunda, Inheritance'in nasıl yapılacağını görelim. 

```python
class BankAccount: 
    
    def __init__(self, balance):
        self.balance = balance
        
    def withdraw(self, amount): 
        self.balance -= amount

class SavingsAccount(BankAccount):
    pass

savings_acct = SavingsAccount(1000)
type(savings_acct)
```

```python
print(savings_acct.balance)
```

```python
savings_acct.withdraw(300)
```

+ Alt sınıfın parantezinin içine, üst sınıfın adı yazılır. Böylece alt sınıf, üst sınıftaki özellik ve metotları da içermiş olur. SavingsAccount adlı alt sınıfın herhangi bir ayrı özelliği ya da metodu olmamasına rağmen nesnesini oluşturabilir ve BankAccount'taki özniteliklerini döndürebiliriz. 

Aşağıdaki kod bloğunda, Employee sınıfına bir alt sınıf oluşturalım.

```python
class Employee:
    MIN_SALARY = 30000
    
    def __init__(self, name, salary = MIN_SALARY):
        self.name = name
        if salary >= Employee.MIN_SALARY:
            self.salary = salary
        else:
            self.salary = Employee.MIN_SALARY
            
    def give_raise(self, amount):
        self.salary = self.salary + amount
        

class Manager(Employee):
    
    def display(self):
        print("Manager", self.name)

mng = Manager("Debbie Lashko", 86500)
print(mng.name)
mng.display()
```

### < Customizing Functionality Via Inheritance (Kalıtım Yoluyla İşlevselliği Özelleştirme) > 

Aşağıdaki kod bloğunda, alt sınıfı daha özelleştirelim.

```python
class BankAccount:
    
    def __init__(self, balance):
        self.balance = balance
        
    def withdraw(self, amount):
        self.balance = self.balance - amount
        
class SavingsAccount(BankAccount):
    
    def __init__(self, balance, interest_rate):
        BankAccount.__init__(self, balance) #1
        self.interest_rate = interest_rate
        
    def compute_interest(self, n_preiods = 1):
        return self.balance * ((1 + self.interest_rate) ** n_periods -1)
    
class CheckingAccount(BankAccount):
    
    def __init__(self, balance, limit):
        BankAccount.__init__(self, balance)
        self.limit = limit
        
    def deposit(self, amount):
        self.balance = self.balance + amount
        
    def withdraw(self, amount, fee = 0):
        if fee <= self.limit:
            BankAccount.withdraw(self, amount - fee)
            
        else:
            BankAccount.withdraw(self, amount - self.limit)
```

```python
acct = SavingsAccount(1000, 0.03)
acct.interest_rate
```

```python
check_acct = CheckingAccount(1000, 25)
check_acct.withdraw(200)
```

+ 1-) SavingAccount hesabının nesneleri, BankAccount sınıfının __ init __ yöntemini çağırarak üst sınıf olan BankAccount sınıfının özniteliklerine de sahip olmuş oldu. Yani SavingsAccount'a girilecek olan balance değeri, BankAccount'un da balance değeri olmuş olacak.

### < Designing for Inheritance and Polymorphism (Kalıtım ve Polimorfizm için Tasarım) > 

**Polimorphism (Çok Biçimlilik):** Farklı sınıflardaki nesneler üzerinde çalışmak için birleşik bir arayüz kullanma anlamına gelir. Aşağıdaki görselde olduğu gibi: 



BankAccount sınıfından miras alan (inheritance) SavingsAccount ve CheckingAccount sınıflarının ikisinde de para çekme yöntemi var. Ancak ChechkingAccount sınıfının withdraw() işlevi farklı kod yürütüyor.

Yani **Polimorphism**, methodların objeye göre farklı çıktılar üretmesi veya farklı işler yapmasıdır. Yani alışageldiğimiz gibi methodlara sabit görevler vermek yerine onlara çok biçimli (polimorf) davranacak şekilde bir esneklik vermektir.

Diyelim ki tüm hesap listesinden aynı miktarda parayı bir kered çekmek için bir fonksiyon tanımladık.

```python
class BankAccount:
    
    def __init__(self, balance):
        self.balance = balance
        
    def withdraw(self, amount):
        self.balance = self.balance - amount
        
class SavingsAccount(BankAccount):
    
    def __init__(self, balance, interest_rate):
        BankAccount.__init__(self, balance) #1
        self.interest_rate = interest_rate
        
    def compute_interest(self, n_preiods = 1):
        return self.balance * ((1 + self.interest_rate) ** n_periods -1)
    
class CheckingAccount(BankAccount):
    
    def __init__(self, balance, limit):
        BankAccount.__init__(self, balance)
        self.limit = limit
        
    def deposit(self, amount):
        self.balance = self.balance + amount
        
    def withdraw(self, amount, fee = 0):
        if fee <= self.limit:
            BankAccount.withdraw(self, amount - fee)
            
        else:
            BankAccount.withdraw(self, amount - self.limit)
```

```python
def batch_withdraw(list_of_accounts, amount):
    for acct in list_of_accounts:
        acct.withdraw(amount)
```

+ Yukarıdaki fonksiyon, kendisine iletilen nesnelerin çek hesapları (check accounts), tasarruf hesapları (saving accounts) olup olmadığını bilmez veya umursamaz. Önemli olan, tek bir argümanı kabul eden para çekme yöntemi sahip olmalarıdır.

```python
b = BankAccount(1000)
c = CheckingAccount(2000, limit = 50000)
s = SavingsAccount(3000, interest_rate = 0.1)

batch_withdraw([b, c, s], 1000)

print(b.balance)
print(c.balance)
print(s.balance)
```

+ Görüleceği üzere, batch_withdraw işlevi b, c ve s nesnelerinden 1000 birim parayı aynı anda çekmiş oldu.

### < Managing Data Access : Private Attributes (Veri Erişimini Yönetme: Özel Nitelikler) > 

Python'daki tüm sınıf verileri teknik olarak herkese açıktır. Herhangi bir sınıfın herhangi bir özniteliği veya yöntemine herkes erişebilir. Verilere erişimi yöntemenin birkaç yolu vardır: 

+ Naming Conventions (Verilerin harici tüketim olmadığını belirtmek için bazı evrensel adlandırma kurallarını kullanmak)
+ Her bir özelliğin nasıl değiştirildiğini kontrol etmemize izin veren özel **@property** adı verilen nitelik türlerini kullanmak

**Naming Conventions Kuralları**

+ İlk ve en önemli kural, bir durumu belirtmek için önde gelen, tek bir alt çizgi ( _ ) kullanmaktır. Bir özniteliğin değerinin geçerliliğini kontrol eden ancak sınıfın bir parçası olmayan ve önceden haber verilmeden değiştirilebilen öznitelik veya yöntemdir.

+ Önde gelen, çift alt çizgi ( __ ) kullanmaktır. Bu şekilde başlayan öznitelik veya yöntemler, bu verinin miras alınmadığı anlamına gelir. Geçersiz kılınmaması gereken önemli nitelikleri ve yöntemleri korumak için önde gelen çift alt çizgi kullanılabilir. Sınıfın mevcut sürümünü depolayan ve kendi sürümlerine sahip olacak alt sınıflarına aktarılmaması gereken bir 'sürüm' özelliğidir. Yani alt sınıflara aktarılmayacak öznitelik ve yöntemleri ifade ediyor.   

Aşağıdaki kod bloğunda, Naming Conventions'u kullanalım. Bir BetterDate sınıfı oluşturacağız ve bu sınıfa tarihin geçerliliğini kontrol eden bir yöntem eklemeyeceğiz ancak bunu BetterDate'in genel arayüzünün bir parçası yapmayacağız.

```python
class BetterDate: 
    
    _MAX_DAYS = 30 #1 
    __MAX_MONTH = 12 #2
    
    def __init__(self, year, month, day):
        self.year = year
        self.month = month
        self.day = day
        
    @classmethod
    def from_str(cls, datestr):
        year, month, day = map(int, datestr.split("-"))
        return cls(year, month, day)
    
    def _is_valid(self): #3
        if self.day <= BetterDate._MAX_DAYS and self.month <= BetterDate.__MAX_MONTH:
            return True
        else:
            return False
        
bd1 = BetterDate(2020, 4, 30)
print(bd1._is_valid())

bd2 = BetterDate(2020, 6, 45)
print(bd2._is_valid())
```

+ 1-) Sınıfın bir parçası olmayan öznitelik
+ 2-) Alt sınıfa aktarılmayacak olan sınıfa ait bir öznitelik
+ 3-) Sınıfın bir parçası olamayan yöntem

Pandas'ta olduğu gibi benzer davranışı uygulamak için **@property** dekaratörünü kullanabiliriz.

```python
class Employer:
    
    def __init__(self, name, new_salary):
        self._salary = new_salary #1
        
    @property #2
    def salary(self):
        return self._salary
    
    @salary.setter #3 
    def salary(self, new_salary):
        if new_salary < 0:
            raise ValueError("Invalid salary")
        self._salary = new_salary

        
emp = Employer("Miriam Azari", 35000)
print(emp.salary) # <-- @property metodunu çalıştırır
emp.salary = 60000 # <-- @salary.setter'ı çalıştırır
print(emp.salary)
emp.salary = -1000 # <-- Invalid salary hatası verir
```

+ 1-) Verileri depolayacak bir 'dahili' öznitelik tanımladık. **_salary** şeklinde yazdık. 
+ 2-) **@property**'yi; adı, kısıtlı özniteliğin (**_salary**) adı olan bir yöntemde kullandık. Daha sonra dahili özniteliği döndürdük. **@property**, herhangi bir alıcı yöntemi kullanmadan özel bir özniteliğin değerini almak için kullanılır. Özel değişkeni döndürdüğümüz yöntemin önüne **@property** satırı koymalıyız. Python aslında **@property**'yi görünce, o değişkenin özel bir değişken olduğunu anlıyor.
+ 3-) Özel değişkenin değerini ayarlamak için yöntemin önünde **@method_name.setter** kullanırız. Setter olarak kullanmalıyız. Biz burada **@salary.setter** kullandık. Çünkü özel değişkenimiz salary'idi. 

Aşağıdaki kod bloğunda, Customer sınıfı için @property'yi uygulayalım.

```python
class Customer: 
    
    def __init__(self, name, new_bal):
        self.name = name 
        if new_bal < 0:
            raise ValueError()
        else:
            self._balance = new_bal
            
    @property
    def balance(self):
        return self._balance
    
    @balance.setter
    def balance(self, new_bal):
        if new_bal < 0:
            raise ValueError("Invalid Balance!")
        else:
            self._balance = new_bal
            print("Setter Method Called")
            
cust = Customer("oulcan", 2000)
cust.balance = 3000
print(cust.balance)
cust.balance = -1000
```
